```python
#passaggio 1
import pandas as pd
import matplotlib.pyplot as plt

#Passaggio 2
path = 'prova.csv'
dataset = pd.read_csv(path)

#Passaggio 3 
print(dataset.head())
print(dataset.describe())
```

```python
# Passaggio 4
colonna1 = dataset['prezzo_medio']
colonna2 = dataset['vendite']

# Passaggio 5
prezzo_medio = dataset['prezzo_medio'] 
vendita = dataset['vendite']
plt.scatter(prezzo_medio, vendita, s=10, c='blue', alpha=0.5)
plt.xlabel('prezzo medio')
plt.ylabel('vendita')
plt.title('Grafico a dispersione')
plt.show()
```

```python
# Passaggio 6
corr_coeff = colonna1.corr(colonna2)
print("Il coefficiente di correlazione è:", corr_coeff)
```

```python
# Passaggio 7
plt.hist(colonna1, bins=10)
plt.xlabel('Prezzo Medio')
plt.ylabel('Conteggio')
plt.title('Istogramma')
plt.show()
```

```python
# Passaggio 8
plt.boxplot(colonna2)
plt.ylabel('Vendite')
plt.title('Boxplot')
plt.show()
```

```python
# Passaggio 9
from matplotlib import dates as mpl_dates
data = dataset['data']
vendita = dataset['vendite']
plt.figure(figsize=(10, 6))
plt.plot(data, vendita, color='blue', marker='o', linestyle='-')
plt.xlabel('Data')
plt.ylabel('vendite')
plt.title('Grafico a linee')
plt.xticks(rotation=45)
plt.tight_layout() 
plt.show()
```

```python
# Passaggio 10
colonna1 = dataset['prezzo_medio']
colonna2 = dataset['vendite']

plt.bar(colonna2.unique(), colonna2.value_counts())
plt.xlabel('vendite')
plt.ylabel('Conteggio')
plt.title('Grafico a barre')
plt.show()
```
